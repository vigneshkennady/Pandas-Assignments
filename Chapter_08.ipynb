```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_url = '/content/Datasets/Ames Housing.csv'
data_frame = pd.read_csv(file_url)
```

```python
data_frame.head(10)
```

```python
data_frame.sale_price.plot(kind = 'hist');
```

```python
data_frame.lot_area.plot(kind = 'hist');
```

```python
data_frame.plot(kind='scatter', y = 'sale_price', x = 'lot_area');
```

```python
data_frame.boxplot(by='overall_condition', column='sale_price');
```

```python
data_frame.plot(kind='scatter', x = 'overall_condition', y = 'sale_price');
```

```python
data_frame.boxplot(by='overall_quality', column='sale_price');
```

```python
data_frame.plot(kind='scatter', x = 'overall_quality', y = 'sale_price');
```

```python
data_frame.boxplot(by='overall_quality', column='year_built');
```

```python
pd.pivot_table(data_frame,index=['neighborhood']).sale_price.plot(kind='barh');
```

```python
pd.pivot_table(data_frame,index=['neighborhood']).overall_quality.plot(kind='barh');
```

```python
pd.pivot_table(data_frame,index=['year_built']).sale_price.plot();
```

```python
pd.pivot_table(data_frame,index=['year_sold']).sale_price.plot();
```

```python
data_frame.plot(kind='scatter', y = 'sale_price', x = 'overall_condition', c = 'overall_quality',colormap='Blues');
```
